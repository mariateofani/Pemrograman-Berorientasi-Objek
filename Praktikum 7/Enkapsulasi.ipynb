```python
class RekeningBank:
  def __init__(self, saldo_awal):
    self.__saldo = saldo_awal

  def get_saldo(self):
    return self.__saldo

  def set_saldo(self, jumlah):
    if jumlah < 0:
     self.__saldo = jumlah
    else:
      print("Jumlah saldo tidak boleh negatif")

  def deposit(self, jumlah):
    if jumlah > 0:
      self.__saldo += jumlah
      print(f"Berhasil deposit: {jumlah}")
    else:
      return("Jumlah deposit harus lebih dari 0")

  def tarik(self, jumlah):
    if jumlah > self.__saldo:
      print("Saldo tidak cukup!")
    elif jumlah <= 0:
      print("Jumlah penarikan harus lebih dari 0")
    else:
      self.__saldo -= jumlah
      return(f"Berhasil tarik: {jumlah}")

  def transfer(self, rekening_tujuan, jumlah):
    biaya_admin = 2500
    total = jumlah + biaya_admin
    if total <= self.__saldo:
      self.__saldo -= total
      rekening_tujuan.deposit(jumlah)
      print(f"Berhasil transfer: {jumlah}")
      print(f"Biaya admin: {biaya_admin}")
    else:
      print("Saldo tidak cukup!")

  def minimum_saldo(self):
    if self.__saldo < 100000:
      return True
    else:
      return False


rekening1 = RekeningBank(1000000)
rekening1.deposit(500000)
rekening1.tarik(200000)
print(f"Saldo rekening1 sekarang: {rekening1.get_saldo()}")

rekening2 = RekeningBank(500000)
rekening1.transfer(rekening2, 500000)
print(f"Saldo rekening1 setelah transfer: {rekening1.get_saldo()}")
print(f"Saldo rekening2 setelah transfer: {rekening2.get_saldo()}")
```

```python
class Mahasiswa:
  def __init__(self, nama, nilai):
    self.__nama = nama
    self.__nilai = nilai

  def get_nama(self):
    return self.__nama

  def get_nilai(self):
    return self.__nilai

  def set_nilai(self, nilai):
    if 0 <= nilai <= 100:
      self.__nilai = nilai
    else:
      print("Nilai harus antara 0-100")
      self.__nilai = 0

  def hitung_grade(self):
    if self.__nilai >= 85:
      return "A"
    elif self.__nilai >= 70:
      return "B"
    elif self.__nilai >= 60:
      return "C"
    else:
      return "D"

  def status_kelulusan(self):
    if self.__nilai >= 60:
      return "Lulus"
    else:
      return "Tidak Lulus"

  def tampilkan_info(self):
    print(f"Nama: {self.__nama}")
    print(f"Nilai: {self.__nilai}")
    print(f"Grade: {self.hitung_grade()}")
    print(f"Status Kelulusan: {self.status_kelulusan()}")

mhs1 = Mahasiswa("Fadil", 90)
mhs2 = Mahasiswa("Budi", 55)
mhs3 = Mahasiswa("Siti", 75)

mhs1.tampilkan_info()
print("-" * 30)
mhs2.tampilkan_info()
print("-" * 30)
mhs3.tampilkan_info()
```

```python
import csv

class Mahasiswa:
  def __init__(self, nama, nilai):
    self.__nama = nama
    self.__nilai = nilai

  def get_nama(self):
    return self.__nama

  def get_nilai(self):
    return self.__nilai

  def set_nilai(self, nilai):
    if 0 <= nilai <= 100:
      self.__nilai = nilai
    else:
      print("Nilai harus antara 0-100")
      self.__nilai = 0

  def hitung_grade(self):
    if self.__nilai >= 85:
      return "A"
    elif self.__nilai >= 70:
      return "B"
    elif self.__nilai >= 60:
      return "C"
    else:
      return "D"

  def status_kelulusan(self):
    if self.__nilai >= 60:
      return "Lulus"
    else:
      return "Tidak Lulus"

  def tampilkan_info(self):
    print(f"Nama: {self.__nama}")
    print(f"Nilai: {self.__nilai}")
    print(f"Grade: {self.hitung_grade()}")
    print(f"Status Kelulusan: {self.status_kelulusan()}")

  @staticmethod
  def rata_rata(daftar_mahasiswa):
    total = sum(m.get_nilai() for m in daftar_mahasiswa)
    rata_rata_nilai = total / len(daftar_mahasiswa)
    return rata_rata_nilai

def rangking_mahasiswa(daftar):
    urut = sorted(daftar, key=lambda m: m.get_nilai(), reverse=True)
    print("\n=== Ranking Mahasiswa ===")
    for i, m in enumerate(urut, start=1):
        print(f"{i}. {m.get_nama()} - {m.get_nilai()}")

def simpan_ke_csv(daftar, nama_file="data_mahasiswa.csv"):
    with open(nama_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Nama", "Nilai", "Grade", "Status"])

        for m in daftar:
            writer.writerow([
                m.get_nama(),
                m.get_nilai(),
                m.hitung_grade(),
                m.status_kelulusan()
            ])
    print(f"\nData berhasil disimpan ke {nama_file}")

mhs1 = Mahasiswa("Fadil", 90)
mhs2 = Mahasiswa("Budi", 55)
mhs3 = Mahasiswa("Siti", 75)

mhs1.tampilkan_info()
print("-" * 30)
mhs2.tampilkan_info()
print("-" * 30)
mhs3.tampilkan_info()
print("-" * 30)

daftar_mahasiswa = [mhs1, mhs2, mhs3]
print(f"Rata-rata nilai mahasiswa: {Mahasiswa.rata_rata(daftar_mahasiswa)}")
rangking_mahasiswa(daftar_mahasiswa)
simpan_ke_csv(daftar_mahasiswa)
```
