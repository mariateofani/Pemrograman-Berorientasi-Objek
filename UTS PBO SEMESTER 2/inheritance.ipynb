```python
# class induk
class Orang:  # mendefinisikan class induk bernama Orang (parent class)
    def __init__(self, nama, alamat, no_hp):  # untuk inisialisasi data orang
        self.nama = nama  # menyimpan atribut nama
        self.alamat = alamat  # menyimpan atribut alamat
        self.no_hp = no_hp  # menyimpan atribut nomor hp

    def tampil_data(self):  # method untuk menampilkan data orang
        return f"Nama: {self.nama}, Alamat: {self.alamat}, No HP: {self.no_hp}"  # mengembalikan string data orang


# class pasien
class Pasien(Orang):  # class pasien turunan dari orang (inheritance)
    def __init__(self, nama, alamat, no_hp, id_pasien):  # untuk inisialisasi data pasien
        super().__init__(nama, alamat, no_hp)  # memanggil constructor dari class orang
        self.id_pasien = id_pasien  # menyimpan id pasien
        self.riwayat_kunjungan = []  # list kosong untuk menyimpan riwayat kunjungan
        self.hasil_survey = []  # list kosong untuk menyimpan hasil survey

    def isi_survey(self, nilai, komentar):  # method untuk mengisi survey
        data = {  # membuat dictionary untuk menyimpan data survey
            "nilai": nilai,  # menyimpan nilai survey
            "komentar": komentar  # menyimpan komentar survey
        }
        self.hasil_survey.append(data)  # menambahkan data survey ke dalam list

    def tampil_survey(self):  # method untuk menampilkan hasil survey
        print(f"\nData Survey Pasien {self.nama}:")  # menampilkan judul survey
        for i, s in enumerate(self.hasil_survey, 1):  # looping data survey dengan nomor urut
            print(f"  Survey {i}: Nilai={s['nilai']}, Komentar={s['komentar']}")  # menampilkan isi survey

    def tampil_detail(self):  # method untuk menampilkan detail pasien
        print("\n=== DATA PASIEN ===")  # menampilkan header
        print(f"ID Pasien : {self.id_pasien}")  # menampilkan id pasien
        print(self.tampil_data())  # memanggil method dari class orang

# class karyawan
class Karyawan(Orang):  # class karyawan turunan dari orang
    def __init__(self, nama, alamat, no_hp, id_karyawan, jabatan):  # constructor karyawan
        super().__init__(nama, alamat, no_hp)  # memanggil constructor orang
        self.id_karyawan = id_karyawan  # menyimpan id karyawan
        self.jabatan = jabatan  # menyimpan jabatan karyawan

    def tampil_detail(self):  # method untuk menampilkan detail karyawan
        print(f"{self.nama} | {self.jabatan} | id: {self.id_karyawan}")  # menampilkan data karyawan


# class tenaga medis
class TenagaMedis(Karyawan):  # class tenaga medis turunan dari karyawan
    def __init__(self, nama, alamat, no_hp, id_karyawan, jabatan, spesialisasi):  # constructor
        super().__init__(nama, alamat, no_hp, id_karyawan, jabatan)  # memanggil constructor karyawan
        self.spesialisasi = spesialisasi  # menyimpan spesialisasi tenaga medis

# class dokter
class Dokter(TenagaMedis):  # class dokter turunan dari tenaga medis
    def __init__(self, nama, alamat, no_hp, id_karyawan, spesialisasi, sip):  # untuk inisialisasi data dokter
        super().__init__(nama, alamat, no_hp, id_karyawan, "dokter", spesialisasi)  # jabatan otomatis "dokter"
        self.sip = sip  # menyimpan nomor sip (surat izin praktik)

    def periksa_pasien(self, pasien):  # method untuk memeriksa pasien
        print(f"Dokter {self.nama} memeriksa pasien {pasien.nama}")  # menampilkan aktivitas pemeriksaan
    def tampil_detail(self):  # override method tampil_detail (polimorfisme)
        print(f"{self.nama} | Dokter ({self.spesialisasi}) | SIP: {self.sip}")  # menampilkan detail dokter


# class perawat
class Perawat(TenagaMedis):  # class Perawat turunan dari TenagaMedis
    def __init__(self, nama, alamat, no_hp, id_karyawan):  # untuk inisialisasi data perawat
        super().__init__(nama, alamat, no_hp, id_karyawan, "Perawat", "Umum")  # jabatan perawat, spesialisasi umum

    def tampil_detail(self):  # override method tampil_detail
        print(f"{self.nama} | Perawat")  # menampilkan data perawat


# class non medis
class NonMedis(Karyawan):  # class nonmedis turunan dari karyawan
    def __init__(self, nama, alamat, no_hp, id_karyawan, jabatan):  # untuk inisialisasi data karyawan
        super().__init__(nama, alamat, no_hp, id_karyawan, jabatan)  # memanggil data inisialisasi karyawan


# class instansi
class Instansi:  # class untuk instansi (misalnya klinik)
    def __init__(self, nama, alamat):  # untuk inisialisasi data instansi
        self.nama = nama  # menyimpan nama instansi
        self.alamat = alamat  # menyimpan alamat instansi
        self.daftar_karyawan = []  # list untuk menyimpan semua karyawan

    def tambah_karyawan(self, karyawan):  # method untuk menambahkan karyawan
        self.daftar_karyawan.append(karyawan)  # menambahkan ke list

    def tampil_semua(self):  # method untuk menampilkan semua data instansi
        print("\n=" + "="*29)
        print("        DATA INSTANSI")
        print("="*30)
        print(f"Nama   : {self.nama}")
        print(f"Alamat : {self.alamat}")

        print("\n--- DAFTAR KARYAWAN ---")
        for k in self.daftar_karyawan:
            k.tampil_detail()


# program utama
if __name__ == "__main__":  # menandakan program utama dijalankan langsung
    # instansi
    klinik = Instansi("Klinik Sehat", "Madiun")  # membuat objek instansi klinik

    # karyawan
    dokter1 = Dokter("Dr. Zahra", "Gresik", "08123", "K001", "Umum", "SIP123")  # membuat objek dokter
    perawat1 = Perawat("Nafhah", "Madiun", "08222", "K002")  # membuat objek perawat
    admin1 = NonMedis("Chelsea", " Caruban", "08555", "K003", "Admin")  # membuat objek admin

    klinik.tambah_karyawan(dokter1)  # menambahkan dokter ke instansi
    klinik.tambah_karyawan(perawat1)  # menambahkan perawat ke instansi
    klinik.tambah_karyawan(admin1)  # menambahkan admin ke instansi

    # pasien
    pasien1 = Pasien("Caca", "Madiun", "08333", "P001")  # membuat objek pasien
    pasien2 = Pasien("Rara", "Ponorogo", "08444", "P002") # membuat objek pasien ke dua

    # aktivitas 1
    dokter1.periksa_pasien(pasien1)  # dokter memeriksa pasien
    pasien1.isi_survey(5, "Pelayanan sangat baik")  # pasien mengisi survey

    # aktivitas 2
    dokter1.periksa_pasien(pasien2)  # dokter memeriksa pasien
    pasien2.isi_survey(4, "Dokter ramah")  # pasien kedua mengisi survey

    # output semua data
    klinik.tampil_semua()  # menampilkan data instansi dan karyawan
    pasien1.tampil_detail()  # menampilkan detail pasien
    pasien2.tampil_detail()  # menampilkan detail pasien
    pasien1.tampil_survey()  # menampilkan hasil survey pasien
    pasien2.tampil_survey()  # menampilkan hasil survey pasien
```
