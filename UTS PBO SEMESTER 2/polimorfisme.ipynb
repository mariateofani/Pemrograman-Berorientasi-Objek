```python
# class induk survey
class Survey:  # class induk (parent) untuk semua jenis survey
    def hitung_persentase(self):  # method untuk menghitung persentase (akan dioverride)
        pass  # tidak ada isi (placeholder untuk subclass)

    def tampil_hasil(self):  # method untuk menampilkan hasil (akan dioverride)
        pass  # tidak ada isi


# sub class survey pilihan (sangat baik, baik, cukup dan kurang)
class SurveyPertanyaan(Survey):  # class turunan dari survey
    def __init__(self, pertanyaan):  # untuk inisialisasi data pertanyaan
        self.pertanyaan = pertanyaan  # menyimpan teks pertanyaan
        self.jawaban = []  # list untuk menyimpan jawaban responden

    def tambah_jawaban(self, pilihan):  # method untuk menambahkan jawaban
        self.jawaban.append(pilihan)  # menambahkan pilihan ke list jawaban

    def hitung_persentase(self):  # override method dari class induk
        total = len(self.jawaban)  # menghitung total jawaban
        hasil = {  # untuk menyimpan jumlah tiap kategori
            "Sangat Baik": 0,
            "Baik": 0,
            "Cukup": 0,
            "Kurang": 0
        }

        for j in self.jawaban:  # loop setiap jawaban
            if j in hasil:  # jika jawaban termasuk kategori yang valid
                hasil[j] += 1  # tambah jumlah kategori tersebut

        for k in hasil:  # loop setiap kategori
            hasil[k] = (hasil[k] / total * 100) if total > 0 else 0  # hitung persentase

        return hasil  # mengembalikan hasil dalam bentuk dictionary

    def tampil_hasil(self):  # override method tampil_hasil
        print(f"\nPertanyaan: {self.pertanyaan}")  # menampilkan pertanyaan
        hasil = self.hitung_persentase()  # memanggil method hitung_persentase
        for k, v in hasil.items():  # loop hasil
            print(f"{k}: {v:.2f}%")  # menampilkan kategori dan persentase (2 angka desimal)


# sub class survey pilihan (ya/tidak)
class SurveyYaTidak(Survey):  # class turunan dari survey
    def __init__(self, pertanyaan):  # untuk inisialisasi data pertanyaan
        self.pertanyaan = pertanyaan  # menyimpan pertanyaan
        self.jawaban = []  # list untuk menyimpan jawaban

    def tambah_jawaban(self, pilihan):  # method tambah jawaban
        if pilihan in ["Ya", "Tidak"]:  # validasi hanya boleh ya/tidak
            self.jawaban.append(pilihan)  # tambahkan ke list

    def hitung_persentase(self):  # override method
        total = len(self.jawaban)  # hitung total jawaban
        hasil = {"Ya": 0, "Tidak": 0}  # Inisialisasi hasil

        for j in self.jawaban:  # Loop jawaban
            hasil[j] += 1  # tambah jumlah sesuai jawaban

        for k in hasil:  # loop hasil
            hasil[k] = (hasil[k] / total * 100) if total > 0 else 0  # hitung persentase

        return hasil  # return hasil

    def tampil_hasil(self):  # override method tampil
        print(f"\nPertanyaan (Ya/Tidak): {self.pertanyaan}")  # tampilkan pertanyaan
        hasil = self.hitung_persentase()  # hitung hasil
        for k, v in hasil.items():  # loop hasil
            print(f"{k}: {v:.2f}%")  # tampilkan persentase


# sub class survey skala 1–5 (rating instansi)
class SurveySkala1_5(Survey):  # class turunan dari Survey
    def __init__(self, nama_instansi):  # untuk inisialisasi data instansi
        self.nama_instansi = nama_instansi  # menyimpan nama instansi
        self.jawaban = []  # list untuk menyimpan nilai rating

    def tambah_jawaban(self, nilai):  # method tambah jawaban
        if 1 <= nilai <= 5:  # validasi nilai harus antara 1 sampai 5
            self.jawaban.append(nilai)  # tambahkan ke list

    def hitung_persentase(self):  # override method
        total = len(self.jawaban)  # hitung total jawaban
        hasil = {i: 0 for i in range(1, 6)}  # dictionary nilai 1-5 dengan default 0

        for j in self.jawaban:  # loop jawaban
            hasil[j] += 1  # tambah jumlah sesuai nilai

        for k in hasil:  # loop hasil
            hasil[k] = (hasil[k] / total * 100) if total > 0 else 0  # hitung persentase

        return hasil  # return hasil

    def hitung_rata_rata(self):  # method untuk menghitung rata-rata
        if len(self.jawaban) == 0:  # jika tidak ada jawaban
            return 0  # return 0
        return sum(self.jawaban) / len(self.jawaban)  # hitung rata-rata

    def tampil_hasil(self):  # override method tampil
        print("\n=" + "="*34)  # garis pembatas
        print("   RATING KESELURUHAN INSTANSI")  # Judul
        print("="*35)  # garis pembatas
        print(f"Instansi: {self.nama_instansi}")  # tampilkan nama instansi

        hasil = self.hitung_persentase()  # hitung distribusi nilai
        rata = self.hitung_rata_rata()  # hitung rata-rata

        print("\nPresentase Nilai:")  # judul (menampilkan presentase dari hasil ranting
        for k, v in hasil.items():  # loop hasil
            print(f"Nilai {k}: {v:.2f}%")  # tampilkan persentase tiap nilai

        print(f"\nRata-rata Rating: {rata:.2f} / 5")  # tampilkan rata-rata


# program utama (demo polimorfisme)l
if __name__ == "__main__":  # menandakan program utama

    # buat objek survey berbeda
    p1 = SurveyPertanyaan("Bagaimana pelayanan dokter?")  # objek survey pilihan
    p2 = SurveyYaTidak("Apakah Anda akan kembali berobat?")  # objek survey Ya/Tidak
    rating = SurveySkala1_5("Klinik Sehat")  # objek survey rating

    # simulasi jawaban
    data_p1 = ["Sangat Baik", "Baik", "Cukup", "Sangat Baik", "Kurang"]  # data dummy survey pilihan
    data_p2 = ["Ya", "Ya", "Tidak", "Ya"]  # data dummy Ya/Tidak
    data_rating = [5, 4, 5, 3, 4, 5, 2]  # data dummy rating

    for j in data_p1:  # loop data_p1
        p1.tambah_jawaban(j)  # tambahkan ke survey p1

    for j in data_p2:  # loop data_p2
        p2.tambah_jawaban(j)  # tambahkan ke survey p2

    for j in data_rating:  # loop data_rating
        rating.tambah_jawaban(j)  # tambahkan ke survey rating

    # polimorfisme terjadi di sinil
    daftar_survey = [p1, p2, rating]  # list berisi berbagai objek survey

    for survey in daftar_survey:  # loop semua objek
        survey.tampil_hasil()  # memanggil method yang sama, tapi hasil beda (polimorfisme)
```
