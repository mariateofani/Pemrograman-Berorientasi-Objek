```python
# class induk pasien
class Pasien:  # mendefinisikan class pasien dengan konsep enkapsulasi
    def __init__(self, nama, alamat, no_hp, email, id_pasien):  # untuk inisialisasi data
        # private attribute
        self.__nama = nama  # atribut private nama (tidak bisa diakses langsung dari luar)
        self.__alamat = alamat  # atribut private alamat
        self.__no_hp = no_hp  # atribut private nomor HP
        self.__email = email  # atribut private email

        # public
        self.id_pasien = id_pasien  # atribut public (boleh diakses langsung)
        self.hasil_survey = []  # list untuk menyimpan hasil survey

    # getter (ambil data)
    def get_nama(self):  # method getter untuk mengambil nama
        return self.__nama  # mengembalikan nilai nama

    def get_alamat(self):  # getter alamat
        return self.__alamat  # mengembalikan alamat

    def get_no_hp(self):  # getter nomor HP
        return self.__no_hp  # mengembalikan no HP

    def get_email(self):  # getter email
        return self.__email  # mengembalikan email

    # setter (ubah data + validasi)
    def set_no_hp(self, no_hp):  # setter untuk mengubah no hp
        if no_hp.isdigit():  # validasi: harus berupa angka
            self.__no_hp = no_hp  # jika valid, simpan
        else:
            print("no hp harus angka!")  # jika tidak valid, tampilkan pesan error

    def set_email(self, email):  # setter untuk email
        if "@" in email:  # validasi sederhana: harus ada tanda @
            self.__email = email  # jika valid, simpan
        else:
            print("Email tidak valid!")  # jika tidak valid, tampilkan pesan

    # method lain
    def isi_survey(self, nilai, komentar):  # method untuk menambahkan hasil survey
        self.hasil_survey.append({  # menambahkan dictionary ke list hasil_survey
            "nilai": nilai,  # menyimpan nilai survey
            "komentar": komentar  # menyimpan komentar
        })

    def tampil_data_aman(self):  # method untuk menampilkan data pasien secara aman
        print("\n=== DATA PASIEN (AMAN) ===")  # judul output
        print(f"ID Pasien : {self.id_pasien}")  # menampilkan ID pasien
        print(f"Nama      : {self.__nama}")  # menampilkan nama (akses dari dalam class)
        print(f"Alamat    : {self.__alamat}")  # menampilkan alamat
        print(f"No HP     : {self.__no_hp}")  # menampilkan no HP
        print(f"Email     : {self.__email}")  # menampilkan email

# buat objek pasien
p1 = Pasien("Rita", "Saradan", "08123456789", "rita@gmail.com", "P001")  # membuat objek pasien

# ❌ AKSES LANGSUNG (TIDAK BOLEH)
# print(p1.__nama)  # ERROR → karena atribut private tidak bisa diakses langsung

# ✅ AKSES LEWAT GETTER
print("Nama:", p1.get_nama())  # mengambil nama menggunakan getter

# ✅ UBAH DATA LEWAT SETTER
p1.set_no_hp("081234557623")  # mengubah nomor HP dengan nilai valid
p1.set_email("ritatri@gmail.com")  # mengubah email dengan nilai valid

# ❌ coba input salah
p1.set_email("salah_email")  # email tidak valid karena tidak ada "@"

# tampilkan data
p1.tampil_data_aman()  # menampilkan data pasien secara aman melalui method
```
