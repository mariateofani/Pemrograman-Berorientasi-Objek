```python
class Kendaraan:
    def __init__(self, kecepatan):
        self.kecepatan = kecepatan

    def berjalan(self):
        print("Kendaraan sedang berjalan")

    def laju(self):
        print(f"Dengan kecepatan {self.kecepatan}")


class Pesawat(Kendaraan):
    def berjalan(self):
        print(f"Pesawat terbang di udara dengan kecepatan {self.kecepatan} km/jam")


class Mobil(Kendaraan):
    def berjalan(self):
        print(f"Mobil berjalan di jalan raya dengan kecepatan {self.kecepatan} km/jam")


class Motor(Kendaraan):
    def berjalan(self):
        print(f"Motor berjalan di jalan raya dengan kecepatan {self.kecepatan} km/jam")


class Kapal(Kendaraan):
    def berjalan(self):
        print(f"Kapal berlayar di laut dengan kecepatan {self.kecepatan} km/jam")


# Membuat objek dengan kecepatan masing-masing
daftar_kendaraan = [
    Pesawat(800),
    Mobil(120),
    Motor(90),
    Kapal(40)
]

# Loop untuk menampilkan
for kendaraan in daftar_kendaraan:
    kendaraan.berjalan()
```
